```python
import numpy as np
import numpy.random as rnd

import matplotlib.pyplot as plt

# Parameters
nSeed = 10
nObservations = 100000
nBins = 1000
```

```python
rnd.seed(nSeed)
a = []
for i in range(10):
    a.append(rnd.uniform(low=0.0, high=1.0, size=None))
np.array(a)
```

```python
rnd.seed(nSeed)
rnd.uniform(low=0.0, high=1.0, size=10)
```

```python
# Uniform
rnd.seed(nSeed)
a = rnd.uniform(low=0.0, high=1.0, size=nObservations)
x = plt.hist(a, bins=nBins)
```

```python
# Normal
rnd.seed(nSeed)
a = rnd.normal(loc=0.0, scale=1.0, size=nObservations)
x = plt.hist(a, bins=nBins)
```

```python
# Binomial
# Probability of success in a sequence with n > 0 independent events
# where every events has success probability 0 <= p <= 1
rnd.seed(nSeed)
for i in range(1, 10):
    l = rnd.binomial(n=1000, p=i/10, size=nObservations)
    x = plt.hist(l, bins=int(nBins/10), label="p={}".format(i/10))
x = plt.legend(loc="upper right")
```

```python
# Pareto
rnd.seed(nSeed)
a = rnd.pareto(a=1000, size=nObservations)
x = plt.hist(a, bins=nBins)
```
